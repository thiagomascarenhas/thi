Descreva a forma da representa√ß√£o geom√©trica da fun√ß√£o vetorial dada por $F(t) = [t + (t/10)sin(t), t + (t/10)cos(t)]$


```python
# Bibliotecas utilizadas

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Vetor (programa√ß√£o) - um array de uma √∫nica dimens√£o, uma estrutura de dados
# que armazena uma sequ√™ncia de informa√ß√µes, todos do mesmo tipo

# Cria um vetor
t = np.linspace(0, 100, 301)

# Fun√ß√£o dada no exemplo
def F(t):
    return np.array([t+t/10*np.sin(t), t+(t/10.)*np.cos(t)])

# Cria 3 diferentes vetores
t1 = np.linspace(0, 30, 200)
t2 = np.linspace(0, 50, 200)
t3 = np.linspace(0, 90, 200)

# Cria a figura e configura os eixos
fig, axes = plt.subplots(1, 3, figsize=(10,3), sharey=True)

# Nomeia cada eixo
ax1 = axes[0]
ax2 = axes[1]
ax3 = axes[2]

# Plota os gr√°ficos referentes a cada conjunto de valores de um vetor, no qual
# F, a fun√ß√£o, est√° em fun√ß√£o de cada vetor

ax1.plot(F(t1)[0], F(t1)[1], label="$t=30$ s", color="blue")
ax2.plot(F(t2)[0], F(t2)[1], label="$t=50$ s", color="blue")
ax3.plot(F(t3)[0], F(t3)[1], label="$t=90$ s", color="blue")


# Desenha a seta
ax1.quiver(0, 0, F(t1)[0,-1], F(t1)[1,-1],
           scale=1, scale_units="xy", width=0.02)

ax2.quiver(0, 0, F(t2)[0,-1], F(t2)[1,-1],
           scale=1, scale_units="xy", width=0.02)

ax3.quiver(0, 0, F(t3)[0,-1], F(t3)[1,-1],
           scale=1, scale_units="xy", width=0.02)

# Determina os limites de cada figura
for ax in axes:
    ax.set_xlim(0,100)
    ax.set_ylim(0,100)
    ax.legend(loc="upper left")

# Nomeia os eixos
ax1.set_ylabel("y")
ax2.set_xlabel("x")

# Exibe a figura
plt.show()
```

#Campos Vetoriais Tridimensionais

Um campo vetorial em $R^3$ √© uma fun√ß√£o $F : D ‚Üí R^3$
, D ‚àà $R^3$
.
Neste caso, o campo vetorial pode ser escrito em termos de
suas componentes $P, Q $ e $ R$ da seguinte forma:

$\vec{F}(x, y, z)= P(x, y, z)\vec{i} + Q(x, y, z)\vec{j} + R(x, y, z)\vec{k}$

=

$(P(x, y, z), Q(x, y, z), R(x, y, z))$



Observe que $P, Q $ e $ R $ s√£o campos escalares, ou seja, fun√ß√µes
de tr√™s vari√°veis e $\vec{i},\vec{j}, \vec{k}$ s√£o os vetores unit√°rios.

Pensando matematicamente e resumidamente sobre o que √© um campo vetorial tridimensional, cada ponto no espa√ßo tridimensional est√° associado a um vetor tridimensional. Podemos visualizar isso como uma fun√ß√£o vetorial (multivari√°vel), cuja entrada √© um ponto $(x, y, z)$ em um espa√ßo tridimensional, e cuja sa√≠da √© um vetor tridimensional.

Um exemplo √© o campo vetorial $\vec{F}=y\vec{i}+z\vec{j}+x\vec{k}$, o qual possui a seguinte representa√ß√£o em alguns pontos de $R^3$:

image.png

###Exemplo:

N√≥s definimos um campo vetorial em $E^3(ou$ $R^3)$ que possui a seguinte forma
$[‚àíy,x,sin(xyz)]$ utilizando os vetores unit√°rios  $\vec{i},\vec{j}, \vec{k}$. Assim, ele se transforma em $\vec{v}=‚àíy\vec{i} +x\vec{j}+sin(xyz)\vec{k}$.


```python
# Bibliotecas utilizadas
import sympy as sp

# Criamos simbolos para representar 'x', 'y' e 'z'
x = sp.symbols('x')
y = sp.symbols('y')
z = sp.symbols('z')

# Definimos uma fun√ß√£o f(x) em R¬≥
vector = [-y, x, sp.sin(x * y * z)]

# Imprimimos a fun√ß√£o
print(vector)
```

Podemos acessar as componentes deste atrav√©s do operador [ ]. Se quisermos o componente da coordenada $x$, por exemplo, basta digitar:

```python
vector[0]
```

J√° se quisermos acessar todas as componentes do vetor, digita-se:

```python
vector[:]
```

O campo vetorial pode ser definido em qualquer ponto $p$ de $E^3(R^3)$. Chamando $p$ o ponto (3, -2, 1), teremos:

```python
# Alteramos os valores de (x,y,z) para (3,-2,1)
valor_x = 3
valor_y = -2
valor_z = 1

# Substituimos os valores no vetor
p = [vector[0].subs({x: valor_x, y: valor_y, z: valor_z}),
             vector[1].subs({x: valor_x, y: valor_y, z: valor_z}),
             vector[2].subs({x: valor_x, y: valor_y, z: valor_z})]

# Imprimimos o resultado
print(p)
```

##Vetor Gradiente

###Defini√ß√£o

No c√°lculo vetorial, o Gradiente (ou Vetor Gradiente) √© um vetor que indica o sentido e a dire√ß√£o na qual, por deslocamento a partir do ponto especificado, obt√©m-se o maior incremento poss√≠vel no valor de uma grandeza a partir da qual se define um campo escalar para o espa√ßo em considera√ß√£o. Constr√≥i-se assim, a partir do campo escalar e de um operador denominado operador gradiente, um campo vetorial, que atrela a cada ponto do espa√ßo o correspondente vetor gradiente para a grandeza em considera√ß√£o.

Em suma, o Vetor Gradiente √© um vetor normal √† curva de n√≠vel no ponto e indica a dire√ß√£o e sentido que dever√° ser deslocada, a partir do ponto especificado, para que haja o maior incremento na grandeza medida. Sua nota√ß√£o √© $‚àáf$, e ele cont√©m todas as informa√ß√µes sobre as derivadas parciais de uma fun√ß√£o multivari√°vel em um vetor.

image.png

Visualizando melhor uma aplica√ß√£o do Gradiente √© s√≥ imaginar algu√©m parado no ponto ($x_0$, $y_0...$) em um sistema de coordenadas no dom√≠nio de $f$, por exemplo. O vetor $‚àáf(x_0, y_0...)$ diria em que dire√ß√£o essa pessoa deveria caminhar para aumentar o valor de $f$ mais rapidamente.

###Exemplo:

Para calcular o Gradiente de uma fun√ß√£o em um ponto, come√ßamos primeiramente introduzindo uma fun√ß√£o $f(x,y,z)$, nos termos de suas coordenadas cartesianas; neste exemplo, utilizaremos uma fun√ß√£o $f(x,y,z) = (x^2, 2yz, senz )$ n√£o especificada de $(x, y, z)$.

##Laplaciano

O Laplaciano de uma fun√ß√£o $f(x,y,z)$ √© definido do seguinte modo: $ \nabla(\nabla F) = \nabla^2 F$. Temos ent√£o que $
\nabla^2 f = \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} + \frac{\partial^2 f}{\partial z^2}
$

O Laplaciano descreve a curvatura e a varia√ß√£o de uma fun√ß√£o escalar em um espa√ßo tridimensional. No espa√ßo do curso de Engenharia de Telecomunica√ß√µes, n√≥s usamos desse conceito para calcular as equa√ß√µes de Laplace e Poisson vistas no Eletromagnetismo Aplicado. No mais o conceito de Laplaciano de uma fun√ß√£o pode ser usada quando usado para resolver equa√ß√µes diferenciais parciais e analisar fen√¥menos de difus√£o, transfer√™ncia de calor e propaga√ß√£o de ondas, entre outros.

###Exemplo:
Se definirmos $f(x,y,z) = x^2 + y^2 + z^2$ e quisermos achar o laplaciano dessa fun√ß√£o, temos:
$
\nabla^2 f = \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} + \frac{\partial^2 f}{\partial z^2}
$
Calculamos separadamente as derivadas parciais de $x, y, z$, ficamos ent√£o com:

$\frac{\partial^2 f}{\partial x^2} = 2$,
$\frac{\partial^2 f}{\partial y^2} = 2$,
$\frac{\partial^2 f}{\partial z^2} = 2$

Com todos os valores, temos que:
$
\nabla^2 f = \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} + \frac{\partial^2 f}{\partial z^2} = 2 + 2 + 2 = 6
$

```python
import sympy as sp

# Definimos as v√°riaveis usada no c√≥digo
x, y, z = sp.symbols('x y z')

# Definimos a fun√ß√£o f(x,y,z)
f = x**2 + y**2 + z**2

# Calculamos o Laplaciano de f
laplaciano = sp.diff(f, x, 2) + sp.diff(f, y, 2) + sp.diff(f, z, 2)
print("Laplaciano de f: ", laplaciano)
```

##Rotacional

O Rotacional de um campo vetorial √© definido como o produto vetorial do Gradiente pelas componentes do pr√≥prio campo vetorial.



Seja $\vec{F}(x, y, z)= P(x, y, z)\vec{i} + Q(x, y, z)\vec{j} + R(x, y, z)\vec{k}$ um campo vetorial definido em $R^3$ e considere que $P, Q, R$ admitam derivadas parciais.
O rotacional de $\vec{F}$, o qual √© denotado por $rot\vec{F}$, √© dado por:

$\nabla \times \mathbf{F}$ = \

\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\
P & Q & R
\end{vmatrix}






Onde $‚àá$ √© o Gradiente de $\vec{F}$.

Para ilustrar uma interpreta√ß√£o do rotacional de um campo vetorial, considere ainda $\vec{F}
$
 ; pense que este √© o campo de velocidade de um fluido e, al√©m disso, uma part√≠cula situada no ponto $(x,y,z)$.

As part√≠culas situadas numa vizinhan√ßa deste ponto, tendem a se movimentar ao redor do eixo formado pelo vetor $rot{\vec{F}}$ e o comprimento deste vetor representa a velocidade com que as part√≠culas se movem ao redor deste eixo.

Se $rot{\vec{F}} = \vec{0}$,
  ent√£o o flu√≠do est√° livre de rota√ß√µes na vizinha do ponto $(x,y,z)$.

###Exemplo:

O Rotacional de um campo vetorial $\vec{u}=xz\vec{i}
‚Äã
 +yz^2\vec{j}
‚Äã
 +x^2y\vec{k}$  √© definido como:

```python
import sympy as sp

# Defina x, y, z como vari√°veis
x, y, z = sp.symbols('x y z')

# Defina o campo vetorial F = (xz, yz^2, x^2y)
Fx = x * z
Fy = y * z**2
Fz = x**2 * y

# Calcule o rotacional de F
rotational_F = (
    sp.diff(Fz, y) - sp.diff(Fy, z),
    sp.diff(Fx, z) - sp.diff(Fz, x),
    sp.diff(Fy, x) - sp.diff(Fx, y)
)

# Imprima os valores
print("Campo Vetorial F:", (Fx, Fy, Fz))
print("Rotacional de F:", rotational_F)
```

```python
# Crie seu pr√≥prio rotacional
# Para utilizar retire o # das linhas ap√≥s a numera√ß√£o

import sympy as sp

# Define x, y, z como vari√°veis
x, y, z = sp.symbols('x y z')

# Defina o seu campo vetorial F = (Fx, Fy, Fz)

#1
#Fx = x * z
#Fy = y * z**2
#Fz = x**2 * y

# Calcule o rotacional de F
rotational_F = (
    sp.diff(Fz, y) - sp.diff(Fy, z),
    sp.diff(Fx, z) - sp.diff(Fz, x),
    sp.diff(Fy, x) - sp.diff(Fx, y)
)

# Imprima os valores
print("Campo Vetorial F:", (Fx, Fy, Fz))
print("Rotacional de F:", rotational_F)
```

 +2yz\vec{j}
‚Äã
 +sen(z)\vec{k}$, √© dado por:

###Integrais Triplas

Continuamos agora com o mesmo conceito que temos de Integrais Simples e Duplas. A Integral Tripla serve para calcular o "hipervolume" sob uma superf√≠cie tridimensional em um dom√≠nio tridimensional. Quando temos alguma superficie que precisa ser representada a partir de tr√™s coordenadas surge a necessidade de usar coordenada tripla. Um exemplo disso √© o cilindro, onde o mesmo √© representada em raz√£o do raio($r$) e do √¢ngulo ($\theta$) que representa o c√≠rculo que o comp√µe. Al√©m dessas coordenadas temos a coordenada $z$ que ir√° definir a altura que o cilindro ir√° ter.

As regras de integra√ß√£o continuam as mesmas, onde integramos primeiro a integral que est√° dentro para a de fora, um exemplo disso √© a seguinte integral:

$ f(x,y,z) = x^2 + y^2 + z^2 $  sobre o paralelep√≠pedo definido pelos intervalos $x$ de [0, 1], $y$ de [0, 1] e $z$ de [0, 1], temos:

\begin{align*}
\iiint_E (x^2 + y^2 + z^2) \, dV &= \int_0^1 \int_0^1 \int_0^1 (x^2 + y^2 + z^2) \, dz \, dy \, dx
\end{align*}

Come√ßamos integrando em rela√ß√£o a $z$ por comodidade, mas poderiamos come√ßar em rela√ß√£o a qualquer vari√°vel:

\begin{align*}
\int_0^1 (x^2 + y^2 + z^2) \, dz &= x^2z + y^2z + \frac{z^3}{3} \bigg|_0^1 \\
&= x^2 + y^2 + \frac{1}{3}
\end{align*}

Agora, integramos o resultado em rela√ß√£o a $y$:

\begin{align*}
\int_0^1 (x^2 + y^2 + \frac{1}{3}) \, dy &= x^2y + \frac{y^3}{3} + \frac{y}{3} \bigg|_0^1 \\
&= x^2 + \frac{1}{3} + \frac{1}{3} \\
&= x^2 + \frac{2}{3}
\end{align*}

Por fim, integramos o resultado em rela√ß√£o a $x$:

\begin{align*}
\int_0^1 (x^2 + \frac{2}{3}) \, dx &= \frac{x^3}{3} + \frac{2x}{3} \bigg|_0^1 \\
&= \frac{1}{3} + \frac{2}{3} \\
&= 1
\end{align*}

O c√≥digo abaixo ir√° calcular a integral tripla.

# Teorema da Diverg√™ncia

O Teorema da Diverg√™ncia tamb√©m conhecido como o Teorema de Gauss √© um dos principais conhecimentos que usamos no curso de C√°lculo Vetorial, tendo diversas aplica√ß√µes sendo elas no Eletromagnetismo, Mec√¢nica dos Flu√≠dos, Geof√≠sica entre v√°rias outras.

###Defini√ß√£o

Quando temos um campo $\vec{F}=P
‚Äã
 (x,y,z)\vec{i}
‚Äã
 +Q
‚Äã
 (x,y,z)\vec{j}
‚Äã
 +R
‚Äã
 (x,y,z)\vec{k}$, $E$ uma regi√£o s√≥lida simples e $S$ a superficie da fronteira dessa regi√£o que est√° com orienta√ß√£o positiva, o teorema da diverg√™ncia afirma o seguinte:  CodeCogsEqn (1).png

O Teorema afirma que sob as condi√ß√µes dadas, o fluxo de $F$ pela fronteira de $E$ √© igual √† integral tripla da diverg√™ncia de $F$ em $E$, facilitando assim o c√°lculo a ser feito, pois usamos apenas uma integral tripla no lugar de decompor a dupla.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definindo as coordenadas cartesianas (x, y, z) para a esfera
theta, phi = np.linspace(0, np.pi, 40), np.linspace(0, 2 * np.pi, 40)
theta, phi = np.meshgrid(theta, phi)
r = 1  # Raio da esfera

x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)

# Definindo o campo vetorial F = x*i + y*j + z*k
Fx = x
Fy = y
Fz = z

# Calculando a diverg√™ncia do campo vetorial F
divergence = Fx + Fy + Fz

# Calculando o fluxo atrav√©s da superf√≠cie da esfera
flux = np.sum(divergence * r**2 * np.sin(theta))

# Plotando o gr√°fico do campo vetorial
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(x, y, z, Fx, Fy, Fz, length=0.1, normalize=True, color='b', label='Campo Vetorial F')
ax.set_title('Campo Vetorial F')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.legend()
plt.show()

# Exibindo o valor do fluxo calculado
print(f'Fluxo atrav√©s da superf√≠cie da esfera: {flux}')
```

# Teorema de Stokes

Teorema de Stokes mostra uma rela√ß√£o entre uma integral de superf√≠cie sobre uma superf√≠cie $S$ com uma integral em torno da curva da fronteira $S$ (que √© uma curva no espa√ßo).

###Defini√ß√£o

Se definirmos $S$ como sendo uma superf√≠cie orientada, suave por partes e que sua fronteira √© formada por uma  curva $C$ fechada com orienta√ß√£o positiva, teremos que o campo vetorial $F$ que tem seus componentes $x, y, z$ onde apresentam derivadas parciais continuas, chegamos a seguinte equa√ß√£o:

$
\oint_C \mathbf{F} \cdot d\mathbf{r} = \iint_S (\nabla \times \mathbf{F}) \cdot d\mathbf{S}
$

Definimos:

$\oint_C \mathbf{F} \cdot d\mathbf{r}$ como a integral de linha do campo vetorial $F$ sobre a curva $C$;

$ \iint_S (\nabla \times \mathbf{F}) \cdot d\mathbf{S}$ como a integral dupla do rotacional do campo vetorial $F$ sobre a superficie $S$;

A aplica√ß√£o do Teorema de Stokes pode ser observada em campos de estudo do Eletromagnetismo, Din√¢mica dos Flu√≠dos, Topologia, F√≠sica Te√≥rica para a simplifica√ß√£o dos c√°lculos entre diversas outras utilidades.

```python
from sympy import symbols, Matrix, cos, sin, pi
from sympy.vector import CoordSys3D, ParametricRegion, vector_integrate, curl

# Definindo as vari√°veis
x, y, z, t = symbols('x y z t')
C = CoordSys3D('C')

# Definindo o campo vetorial
F = -C.y*C.i + C.x*C.j + C.z*C.k

# Definindo a curva param√©trica do c√≠rculo unit√°rio no plano xy
curve = ParametricRegion((cos(t), sin(t), 0), (t, 0, 2*pi))

# Calculando a integral de linha sobre C
line_integral = vector_integrate(F, curve)

# Calculando o rotacional de F
curl_F = curl(F)

# Definindo a superf√≠cie param√©trica do disco unit√°rio no plano xy
surface = ParametricRegion((x*cos(y), x*sin(y), 0), (x, 0, 1), (y, 0, 2*pi))

# Calculando a integral de superf√≠cie sobre S
surface_integral = vector_integrate(curl_F, surface)

print(f"Integral de linha: {line_integral}")
print(f"Integral de superf√≠cie: {surface_integral}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definindo pontos para o c√≠rculo no plano xy
theta = np.linspace(0, 2*np.pi, 100)
x_curve = np.cos(theta)
y_curve = np.sin(theta)
z_curve = np.zeros_like(theta)

# Definindo pontos para o disco unit√°rio no plano xy
r = np.linspace(0, 1, 100)
theta_grid, r_grid = np.meshgrid(theta, r)
x_surface = r_grid * np.cos(theta_grid)
y_surface = r_grid * np.sin(theta_grid)
z_surface = np.zeros_like(x_surface)

# Criando a figura e o eixo 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotando a curva e a superf√≠cie
ax.plot(x_curve, y_curve, z_curve, label='Curve', color='b')
ax.plot_surface(x_surface, y_surface, z_surface, alpha=0.5)

# Adicionando detalhes ao gr√°fico
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Visualiza√ß√£o da Curva e da Superficie')
ax.legend()

# Mostrando o gr√°fico
plt.show()
```
