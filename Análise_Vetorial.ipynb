# 1) <big> Escalares e Vetores

- **Escalares:**
  - Uma grandeza escalar é representada apenas por um número real.
  - Não possui direção nem sentido.
  - Exemplos: temperatura, massa, tempo, densidade, etc.

- **Vetores:**
  - Uma grandeza que possui magnitude (valor numérico) e direção.
  - Descreve uma quantidade física com tamanho e direção.
  - Usados para representar deslocamentos, velocidades, acelerações, forças, etc.
  - Representados por setas em um gráfico, onde o comprimento da seta representa a magnitude e a direção da seta representa a direção do vetor.

  <br>


# 1.2) <big> Álgebra Vetorial

<big> **Exemplo de Adição de Vetores**

Suponha que tenhamos dois vetores no plano XYZ:

- Vetor $\vec{\mathbf{v}} = [2, 3,1]$
- Vetor $\vec{\mathbf{u}} = [4, 1,1]$

Podemos calcular a soma dos vetores $\vec{\mathbf{v}}$ e $\vec{\mathbf{u}}$:

$$\vec{\mathbf{v}} + \vec{\mathbf{u}} = \begin{bmatrix} 2 \\ 3 \\ 1 \end{bmatrix} + \begin{bmatrix} 4 \\ 1 \\ 1 \end{bmatrix} = \begin{bmatrix} 2 + 4 \\ 3 + 1 \\ 1 + 1 \end{bmatrix} = \begin{bmatrix} 6 \\ 4 \\ 2 \end{bmatrix}$$

<br>

Portanto, a soma dos vetores $\vec{\mathbf{v}}$ e $\vec{\mathbf{u}}$ resulta no vetor $\begin{bmatrix} 6 \\ 4 \\2 \end{bmatrix}$.

<br>

<big> **Exemplo de Multiplicação por Escalar**

Suponha que tenhamos o seguinte vetor e escalar:

- Vetor $\vec{\mathbf{v}} = [2, 3,1]$
- Escalar $k = 2$

<br>

Para calcular a multiplicação do vetor $\vec{\mathbf{v}}$ pelo escalar $k$, multiplicamos cada componente do vetor pelo escalar:

$$\vec{\mathbf{v}} \cdot k = \begin{bmatrix} 2 \\ 3 \\1 \end{bmatrix} \cdot 2$$

Realizando a multiplicação, obtemos:

$$\vec{\mathbf{v}} \cdot k = \begin{bmatrix} 2 \cdot 2 \\ 3 \cdot 2 \\ 1 ⋅ 2 \end{bmatrix} = \begin{bmatrix} 4 \\ 6 \\ 2 \end{bmatrix}$$

<br>

Portanto, a multiplicação do vetor $\vec{\mathbf{v}}$ pelo escalar $k$ resulta no vetor $\begin{bmatrix} 4 \\ 6  \\2 \end{bmatrix}$.



<br>

<big> **Exemplo de Produto Escalar**

Suponha que tenhamos os seguintes vetores:

- Vetor $\vec{\mathbf{v}} = [2, 3,1]$
- Vetor $\vec{\mathbf{u}} = [4, 1,1]$

Para calcular o produto escalar entre dois vetores $\vec{\mathbf{v}}$ e $\vec{\mathbf{u}}$, você pode usar a seguinte fórmula:

$$ \vec{\mathbf{v}} \cdot \vec{\mathbf{u}} = |\vec{\mathbf{v}}| \cdot |\vec{\mathbf{u}}| \cdot \cos(\theta) $$


Se você conhece as componentes dos vetores $ \vec{v} $ e $ \vec{u} $, você pode usar a seguinte fórmula alternativa:

$$ \vec{\mathbf{v}} \cdot \vec{\mathbf{u}} = v_x \cdot u_x + v_y \cdot u_y + v_z \cdot u_z $$

Realizando os produtos e somando, obtemos:

$$\vec{\mathbf{v}} \cdot \vec{\mathbf{u}} = (2 \cdot 4) + (3 \cdot 1) + (1⋅1) = 8 + 3 + 1= 12$$

<br>

Portanto, o produto escalar entre os vetores $\vec{\mathbf{v}}$ e $\vec{\mathbf{u}}$ é igual a 12.

<br>

### <big> **Produto Vetorial (Produto Externo):**

<big> **Exemplo de Produto Vetorial**

Suponha que tenhamos os seguintes vetores:

- Vetor $\vec{\mathbf{v}} = [1, 2,3]$
- Vetor $\vec{\mathbf{u}} = [4, 5,6]$

O produto vetorial entre os vetores $\vec{\mathbf{v}}$ e $\vec{\mathbf{u}}$ em 3D é dado pelo seguinte cálculo:

$$
\vec{\mathbf{u}} \times \vec{\mathbf{v}} = \begin{pmatrix} i & j & k \\ u_1 & u_2 & u_3 \\ v_1 & v_2 & v_3 \end{pmatrix}
$$

Subtituindo os valores, obtemos:

$$
\vec{\mathbf{u}} \times \vec{\mathbf{v}} = \begin{pmatrix} i & j & k \\ 1 & 2 & 3 \\ 4 & 5 & 6 \end{pmatrix}
$$

Aplicando a regra de Sarrus para calcular o determinante desta matriz 3x3, obtemos:

$$\vec{\mathbf{u}} \times \vec{\mathbf{v}} = (2 \cdot 6 - 3 \cdot 5) \mathbf{i} - (1 \cdot 6 - 3 \cdot 4) \mathbf{j} + (1 \cdot 5 - 2 \cdot 4) \mathbf{k}$$

$$\vec{\mathbf{u}} \times \vec{\mathbf{v}} =  -3 \mathbf{i} +6\mathbf{j} -3 \mathbf{k}$$

<br>

Portanto, o produto vetorial de $\vec{\mathbf{u}}$ e $\vec{\mathbf{v}}$ é $[-3, 6, -3]$.

---

<br>

<big> <big> **Regra da mão direita**

A regra da mão direita é fundamental na álgebra vetorial para determinar a orientação dos vetores em operações como o produto vetorial. No produto vetorial, o vetor resultante é perpendicular ao plano definido pelos vetores iniciais, e sua direção é determinada estendendo-se os dedos da mão direita na direção do primeiro vetor e dobrando-os em direção ao segundo vetor.

<center>

<img src="https://media.giphy.com/media/DAaZreMhasM6tWURjx/giphy.gif" alt="regra da mao direita" width="480" height="480">

<small>Fonte: Próprios autores</small>

</center>


Embora a regra da mão direita seja fundamental para determinar a orientação dos vetores no produto vetorial, ela não é aplicável ao produto escalar, pois este último não resulta em um vetor com direção.

<br>

<big> **Representação dos Vetores**

Antes de introduzir os vetores unitários, vamos discutir a representação padrão dos vetores.


Até agora, os vetores foram tratados como matrizes linha, onde cada elemento representa uma coordenada como:

$$[i, j, k]$$

Para integrá-lo em um código, é necessário apresentá-lo na forma computacional. No entanto, quando estamos realizando cálculos vetoriais manualmente, essa abordagem de representação não é muito prática. Uma forma mais conveniente de representar os vetores é através da notação de suas componentes, como por exemplo, $$i + j + k$$.


---

<br>

<big> **Exemplo de Vetor Unitário**

Suponha que tenhamos os seguintes vetores:

- Vetor $\vec{\mathbf{v}} = 2i + 3j -1k  $



Primeiro calculamos a intensidade do vetor $\vec{\mathbf{v}}$:

$$ |\vec{\mathbf{v}}| = \sqrt []{2^2 +3^2 + (-1)^2 } = \sqrt []{14 }  $$

Agora para calculamos o vetor unitário é necessário dividir o vetor pelp seu modulo:

$$ \vec{\mathbf{u}} = \frac{\vec{\mathbf{v}}} {|\vec{\mathbf{v}}|} = \frac{2i + 3j -1k} {\sqrt[]{14}}  = \frac{2i}{ \sqrt[]{14} }+\frac{3j}{ \sqrt[]{14} }-\frac{1k}{ \sqrt[]{14} } $$

$$ \vec{\mathbf{u}} = 0.535i + 0.802j -0.267k $$

<br>

Portanto, o vetor unitário $ \vec{\mathbf{u}} $ resulta em $ 0.535i + 0.802j -0.267k $.

<br>

---



```python
# @title Vetor Unitário {"run":"auto","vertical-output":true}


# @markdown <big> Por favor, forneça valores para aplicar a operação de **Vetor Unitário** de vetores. Por exemplo:

# @markdown - $\vec{\mathbf{v}} = 1i$

# @markdown - $\vec{\mathbf{v}} = 1i + 2j + 3k$

# @markdown - $\vec{\mathbf{v}} = 2i -j $

# @markdown <br>

# @markdown Para integrá-lo em um código, é necessário apresentá-lo na forma computacional, ou seja,

# @markdown - $\vec{\mathbf{v}} =1$

# @markdown - $\vec{\mathbf{v}} = 1,2,3$

# @markdown - $\vec{\mathbf{v}} = 2,-1$

# @markdown <br>



v = [2,3,-1] # @param {type:"raw"}

import numpy as np
def vetor_unitario(vetor):
    modulo = np.linalg.norm(vetor)  # Calcula a magnitude do vetor
    if modulo == 0:
        return vetor
    return vetor / modulo  # Divide o vetor por seu módulo

# Calcula o vetor unitário
vetor_unitario = vetor_unitario(v)

#print("Vetor original:", v)
print(f"O resultado do \033[1m\033[4mVetor Unitário\033[0m é: {vetor_unitario[0]:.3f}i + {vetor_unitario[1]:.3f}j + {vetor_unitario[2]:.3f}" )
```

# <big> 1.4) Sistema de Coordenadas
Um sistema de coordenadas é uma estrutura matemática que permite representar pontos no espaço de forma precisa. Ele é essencialmente um conjunto de regras que descrevem como atribuir números a pontos específicos, facilitando a localização e a descrição de posições.

<br>

## <big> Sistema de Coordenadas Cartesianas

O **Sistema de Coordenadas Cartesianas**, também conhecido como Coordenadas Retangulares, é o sistema mais comumente utilizado em matemática e física. é uma maneira de descrever a posição de um ponto no espaço usando três coordenadas ortogonais: $x$, $y$ e $ z$.

<br>

A representação de coordenadas com vetores $i$, $j$ e $k$ estárelacionada com o sistema de coordenadas retangulares. Onde:

- $i$ representa a unidade de deslocamento ao longo do eixo $x$, apontando na direção do eixo $x$. Dessa forma, podemos escrevê-lo como **$\vec{a}_x$**

- $j$ representa a unidade de deslocamento ao longo do eixo $y$, apontando na direção do eixo $y$. Dessa forma, podemos escrevê-lo como **$\vec{a}_y$**
- $k$ representa a unidade de deslocamento ao longo do eixo $z$, apontando na direção do eixo $z$. Dessa forma, podemos escrevê-lo como **$\vec{a}_z$**

#### <big> Sistema de Coordenadas Cilíndricas

No **Sistema de Coordenadas Cilíndricas**, um ponto no espaço tridimensional é descrito usando as coordenadas $ρ$, $ϕ$ e $z$. Cada uma dessas coordenadas tem um significado específico:

- $ρ$: Representa a distância da origem ao ponto. Dessa forma, podemos escrevê-lo como  $\vec{a}_\rho$.

- $ϕ$: É o ângulo no plano ( $xy$ ) medido a partir do eixo $x$, variando de $0$ a $2π$. Dessa forma, podemos escrevê-lo como  $\vec{a}_\phi$.

- $z$: Indica a altura do ponto em relação ao plano $ $xy$ $. Dessa forma, podemos escrevê-lo como  $\vec{a}_z$.

O sistema de coordenadas cilíndricas é particularmente útil para descrever pontos em situações de simetria cilíndrica, como em cilindros, tubos e outros objetos cilíndricos. <br>

#### <big> Sistema de Coordenadas Esféricas

O **Sistema de Coordenadas Esféricas**, assim como o sistema cartesiano, é uma forma de descrever pontos no espaço tridimensional. Ele utiliza três coordenadas diferentes: $r$, $θ$ e $\phi$. Aqui está a descrição de cada uma delas:

- $r$: Representa a distância do ponto à origem. É a magnitude do vetor posição do ponto. Dessa forma, podemos escrevê-lo como  $\vec{a}_r$

- $θ$: Representa o ângulo no plano $xy$, medido a partir do eixo $x$. Varia de $0$ a $2\pi$ radianos ou de $0^\circ$ a $ 360^\circ$. Dessa forma, podemos escrevê-lo como  $\vec{a}_\theta$
- $\phi$: Representa o ângulo entre o vetor posição do ponto e o eixo $z$. Varia de $0$ a $\pi$ radianos ou de $0^\circ$ a $ 180^\circ$. Dessa forma, podemos escrevê-lo como  $\vec{a}_\phi$

O sistema de coordenadas esféricas é especialmente útil para descrever pontos em torno de uma origem com simetria esférica, como por exemplo em problemas de física, astronomia e engenharia. <br>

## <big> 1.5) Relação entre sistemas de coordenadas

A habilidade de mudar o sistema de coordenadas é fundamental em muitos problemas, especialmente quando lidamos com simetrias cilíndricas ou esféricas.



Se supormos um vetor $\vec{A}$, de componentes cartesianas $(\mathbf{\vec{a}}_x,\mathbf{\vec{a}}_y,\mathbf{\vec{a}}_z)$ e precisamos transformar tal vetor para componentes cilíndricas iremos notar que ele irá ser transformado para $(\mathbf{\vec{a}}_\rho,\mathbf{\vec{a}}_\phi,\mathbf{\vec{a}}_z)$.



Olhando apenas para a relação de transformação referente à $\mathbf{\vec{a}}_\rho$ notamos que ele depende das componentes cartesianas $\mathbf{\vec{a}}_x$ e $\mathbf{\vec{a}}_y$, logo podemos ter casos onde meu $\mathbf{\vec{a}}_x$ é igual meu $\mathbf{\vec{a}}_\rho$, ou casos onde meu $\mathbf{\vec{a}}_y$ é igual meu $\mathbf{\vec{a}}_\rho$.



Para o código abaixo vemos as possíveis direções e sentidos do vetor unitário $a_rho$, ele varia como um círculo, visto que ele é definido para o sistema de coordenadas cilíndricas, como podemos ver abaixo ao variarmos o nosso $a_y$ e $a_x$ temos possíveis valores diferentes de $a_rho$ e veremos isso logo após quando tratamos de transformações de sistemas de coordenadas.

```python
# @title Possíveis direções de $A_\rho$ { run: "auto", vertical-output: true}

import numpy as np
import matplotlib.pyplot as plt

# Função para calcular o vetor a_rho
def a_rho(phi):
    # Vetores unitários nas direções x e y
    a_x = np.array([1, 0])  # Vetor na direção x
    a_y = np.array([0, 1])  # Vetor na direção y

    # Equação para a_rho
    a_rho_vec = np.cos(phi) * a_x + np.sin(phi) * a_y
    return a_rho_vec

# Geração de valores de phi
phi_values = np.linspace(0, 2 * np.pi, 100)

# Plotando os vetores a_rho para diferentes valores de phi
fig, ax = plt.subplots()

for phi in phi_values:
    a_rho_vec = a_rho(phi)
    ax.quiver(0, 0, a_rho_vec[0], a_rho_vec[1], angles='xy', scale_units='xy', scale=1, color='r', alpha=0.3)

# Ajustes da plotagem
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_aspect('equal')
ax.set_xlabel('$a_x$')
ax.set_ylabel('$a_y$')
ax.set_title('Plot dos vetores $a_\\rho$')

plt.grid(True)
plt.show()
```

|       | $\vec{a}_r$ | $\vec{a}_\theta$ | $\vec{a}_\phi$ |

### <big> Exemplo 1.5.1
Dado $G=yz\textbf{a}_x+xz\textbf{a}_y+xy\textbf{a}_z$. Transforme $G$ para coordenadas cilíndricas e esféricas.

Para resolver tal problema podemos utilizar as matrizes de transformação de sistemas de corrdenas, resolvendo para o item $(a)$ ficamos assim: <br>

$$\vec{G}_{car}= [yz \;\;\;\;\; xz \;\;\;\;\; xy] $$


$$\vec{G}_{car2cil}= \begin{vmatrix}
    cos(\phi) & -sen(\phi) & 0\\
    sen(\phi) & cos(\phi) & 0 \\
   0 & 0 & 1
\end{vmatrix}$$
Também transformaremos $x$ e $y$ para o sistema cilindrico:
<br>

$$x= \rho cos(\phi)$$
$$y= \rho sen(\phi)$$
$$z=z$$

$$\vec{G}_{car}= \left[\rho sen(\phi)z \;\;\;\;\; \rho cos(\phi)z \;\;\;\;\; \rho^2 sen(\phi)cos(\phi)\right] $$
Agora iremos multiplicar $\vec{G}_{car}$ com $\vec{G_{car2cil}}$:

$$\vec{G_{cil}}=\left[\rho sen(\phi)z \;\;\;\;\; \rho cos(\phi)z \;\;\;\;\; \rho^2 sen(\phi)cos(\phi)\right]\begin{bmatrix}
    cos(\phi) & -sen(\phi) & 0\\
    sen(\phi) & cos(\phi) & 0 \\
   0 & 0 & 1
\end{bmatrix}$$
$$\vec{G_{cil}}= \left[2pzsin(\phi)cos(\phi) \;\;\;\;\;\;\; -\rho z sen^2(\phi)+\rho z cos(\phi)^2 \;\;\;\;\;\;\; \rho^2 sen(\phi)cos(\phi)\right]$$

----

```python
# @title Item (a)
import sympy as sp

# Definindo as variáveis ou coordenadas
rho, phi, z = sp.symbols('rho phi z')
x = rho * sp.cos(phi)
y = rho * sp.sin(phi)

# Definindo a expressão do campo vetorial ou COMPONENTES DO CAMPO
G_car = sp.Matrix([y*z, x*z, x*y])
#Matriz de transformação
T = sp.Matrix([
[sp.cos(phi), sp.sin(phi), 0],
[-sp.sin(phi), sp.cos(phi), 0],
[0, 0, 1]
])
# Substituindo as coordenadas cilíndricas
G_cil = T*G_car

G_cil
```

### Transformação para Coordenadas Esféricas

As transformações de coordenadas para o sistema esférico são dadas por:

$$x = r \sin(\theta) \cos(\phi)$$
$$y = r \sin(\theta) \sin(\phi)$$
$$z = r \cos(\theta)$$

Agora, substituindo $ x $, $ y $, e $ z $ na expressão de $ G $:

1. **Para $ yz $:**

$$
yz = (r \sin(\theta) \sin(\phi))(r \cos(\theta)) = r^2 \sin(\theta) \sin(\phi) \cos(\theta)
$$

2. **Para $ xz $:**

$$
xz = (r \sin(\theta) \cos(\phi))(r \cos(\theta)) = r^2 \sin(\theta) \cos(\phi) \cos(\theta)
$$

3. **Para $ xy $:**

$$
xy = (r \sin(\theta) \cos(\phi))(r \sin(\theta) \sin(\phi)) = r^2 \sin^2(\theta) \cos(\phi) \sin(\phi)
$$

Portanto, podemos escrever $ G $ em coordenadas esféricas como:

$$
\vec{G}_{esf} = \left[r^2 \sin(\theta) \sin(\phi) \cos(\theta) \; r^2 \sin(\theta) \cos(\phi) \cos(\theta) \; r^2 \sin^2(\theta) \cos(\phi) \sin(\phi)\right]
$$

Agora, organizando a expressão, temos:

$$
\vec{G}_{esf} = \left[ r^2 \sin(\theta) \cos(\theta) \sin(\phi) \; r^2 \sin(\theta) \cos(\theta) \cos(\phi) \; r^2 \sin^2(\theta) \cos(\phi) \sin(\phi) \right]
$$

---

```python
# @title Item (b)
import sympy as sp

# Definindo as variáveis
r, theta, phi = sp.symbols('r theta phi')
x = r * sp.sin(theta) *sp.cos(phi)
y = r * sp.sin(theta) *sp.sin(phi)
z= r * sp.cos(theta)

# Definindo a expressão
G_car = sp.Matrix([y*z, x*z, x*y])
#Matriz de transformação
T = sp.Matrix([
[sp.sin(theta) * sp.cos(phi), sp.cos(theta) * sp.cos(phi),  -sp.sin(phi)],
[-sp.sin(theta) * sp.sin(phi), sp.cos(theta) * sp.sin(phi), sp.cos(phi)],
[sp.cos(theta), -sp.sin(theta), 0]
])
# Substituindo as coordenadas cilíndricas
G_esf = sp.simplify(T*G_car)

G_esf
```

## <big> Integral de Linha
As integrais de linha são basicamente uma integral ao longo de um vetor, tal vetor é o caminho que percorremos ao longo da linha. <br>
A integral de linha pode ser de duas formas:


-   Integral de linha convencional:<br> <br>
$$\int_L \vec{A} \cdot d\vec{l}$$<br>

-   Integral de linha de caminho fechado: <br> <br>
 $$\oint_L \vec{A} \cdot d\vec{l}$$  <br>


Pelo fato de termos um produto escalar na integral, podemos resolvê-la utilizando a relação apresentada anteriormente: $$ \vec{A} \cdot \vec{B}=\left|\vec{A}\right|.\left|\vec{B}\right|cos(\theta).$$
<br>
Porém temos um pequeno problema a partir de agora, como encontramos o nosso $d\vec{l}$? <br>
Devemos saber que o $d\vec{l}$ **NUNCA MUDA**, não importa o sistema de coordenadas que iremos utilizar, sempre ele será o mesmo. <br>

*   Coordenadas cartesianas
$$d\vec{l}= dx \vec{a}_x + dy \vec{a}_y + dz \vec{a}_z$$
*   Coordenadas cilíndricas
$$ d\vec{l}= d\rho \vec{a}_\rho + \rho d\phi \vec{a}_\phi + dz \vec{a}_z$$
*   Coordenadas esféricas
$$ d\vec{l}= dr \vec{a}_r + r d\theta \vec{a}_\theta + r sen(\theta) d\phi \vec{a}_\phi$$

### <big> Exemplo 1.6.1

(a) Calcule a circulação de $\vec{A}= \rho cos(\phi) \vec{a}_\rho + z sen(\phi) \vec{a}_z$ em torno da borda $L$ da fatia definida por $0\leq \rho \leq 2, 0 \leq \phi \leq 60°, z=0 $ e mostrada na imagem abaixo. <br>
<center> <img src='https://i.imgur.com/156YAQA.png'>


Iremos resolver primeiramente da forma convencional e logo após na forma computacional: <br>
O primeiro passo para resolver essa questão é entender que
$$\int_L \vec{A} \cdot d\vec{l} = \left(\int_{L_1} + \int_{L_2} + \int_{L_3}\right)\vec{A} \cdot d\vec{l} = \int_{L_1}\vec{A} \cdot d\vec{l} + \int_{L_2}\vec{A} \cdot d\vec{l} + \int_{L_3}\vec{A} \cdot d\vec{l} $$
Tá, mas o que é o $L_1$, $L_2$ e $L_3$? Basicamente esses são os caminhos que iremos seguir ao longo do trajeto que a figura mostra, o primeiro caminho vai da nossa origem $(0,0,0)$ até nosso $(2,0,0)$, já nosso próximo trajeto vai no final do último $(2,0,0)$ até o nosso $(2,\frac{\pi}{3},0)$ e nosso último trajeto vai do $(2,\frac{\pi}{3},0)$ à $(0,0,0)$. É estranho ter o ângulo no nosso ponto acima, porém estamos trabalhando com coordenadas cilíndricas, então é válido tal notação. <br>
Após entermos nossos trajetos iremos resolver o primeiro passo que é nosso produto escalar $\vec{A} \cdot d\vec{l}$, sabemos que nosso $d\vec{l}$ nunca muda, logo esse produto escalar sempre será o mesmo, então temos que: <br>
$$\vec{A} \cdot d\vec{l}= (\rho cos(\phi) \vec{a}_\rho + zsen(\phi)\vec{a}_z) \cdot (d\rho \vec{a}_\rho + \rho d\phi \vec{a}_\phi + dz \vec{a}_z)$$
$$\vec{A} \cdot d\vec{l} = \rho cos(\phi)d\rho + zsen(\phi)dz$$
Com isso em mente basta agora calcularmos a integral ao longo de cada caminho, então iremos para cada trajeto colocar no lugar do nosso produto escalar nosso resultado de cima e nos limites de integração será do ponto inicial ao ponto final de cada trajeto, ficando assim com:

*   Integral $L_1$
$$\int_{(0,0,0)}^{(2,0,0)} \rho cos(\phi)d\rho + zsen(\phi)dz= \int_{(0,0,0)}^{(2,0,0)}\rho cos(\phi)d\rho+ \int_{(0,0,0)}^{(2,0,0)}zsen(\phi)dz$$
Notamos que o $\phi$ e o $z$ não estão variando.
$$cos(\phi)|_0 \left(\frac{\rho ^2}{2}|_0^2\right) + sen(\phi)|0\left(\frac{z^2}{2}|_0^0\right)$$
$$cos(0)\left(\frac{2^2}{2} -\frac{0 ^2}{2}\right) + sen(0)\left(\frac{0^2}{2} -\frac{0^2}{2}\right)$$
$$1(2)-0(0)=2$$
Logo o resultado da integral do nosso primeiro trajeto é $2$.

*   Integral $L_2$<br>
Para essa integral e para a próxima só precisamos mudar o limite de integração, pois nosso produto escalar é **SEMPRE O MESMO**, o que **MUDA** é o **LIMITE DE INTEGRAÇÃO**, então ficamos com:
$$ \int_{(2,0,0)}^{(2,\frac{\pi}{3},0)}\rho cos(\phi)d\rho+ \int_{(2,0,0)}^{(2,\frac{\pi}{3},0)}zsen(\phi)dz$$
Notamos agora que nosso $\rho$ é fixo, agora o que muda é apenas nosso $\phi$, então resolvendo a integral teremos:
$$cos(\phi) \left(\frac{\rho ^2}{2}|_2^2\right) + sen(\phi)\left(\frac{z^2}{2}|_0^0\right)$$
$$[cos(\phi)](\frac{2^2}{2} -\frac{2^2}{2}) + [sen(\phi)]\left(\frac{0^2}{2} -\frac{0^2}{2}\right) = 0$$ Logo a integral ao longo do segundo trajeto é $0$, pois nem o $\rho$ nem o $z$ estão variando.
*   Integral $L_3$ <br>
Por fim temos nossa terceira integral, em que devemos repetir os mesmos passos, apenas trocando o limite de integração, então ficamos com:
$$\int_{(2,\frac{\pi}{3},0)}^{(0,0,0)}\rho cos(\phi)d\rho+ \int_{(2,\frac{\pi}{3},0)}^{(0,0,0)}zsen(\phi)dz$$
Notamos agora que meu $\rho$ e meu $\phi$ estão variando, porém meu $\phi$ ao longo de todo o trajeto sempre será $\frac{\pi}{3}$, como nosso $z$ não varia, acredito que até aqui já deu para entender que o que está acompanhado do nosso $dz$ da zero pelo fato do $z$ não variar, então temos:
$$cos(\phi)|_\frac{\pi}{3} \left(\frac{\rho ^2}{2}|_2^0\right)=cos\left(\frac{\pi}{3}\right)\left(\frac{0^2}{2}-\frac{2^2}{2}\right)$$
$$\left(\frac{1}{2}\right)(-2)=-1$$ então temos que nossa integral ao longo do terceiro trajeto é $-1$.

Agora vamos somar os três resultados, assim tendo $2+0-1=1$, então a circulação de $\vec{A}$ em torno da borda $L$ é $1$.






```python
# @title Item (a) { run: "auto", vertical-output: true }

#Calculo da Cirlulação
import numpy as np
from scipy.integrate import quad #Quad utiliza métodos computacionais para calcular integrais
#geralmente Quad utiliza quadratura de Gauss para resolver a integral

# Define a função do campo A
def A(p, phi):
    rho = p * np.cos(np.radians(phi))
    z = p * np.sin(np.radians(phi))
    return rho * np.cos(np.radians(phi)), z * np.sin(np.radians(phi))

# Define a função para calcular a circulação
def circulacao():
    # Integral de 0 a 2 de p*cos(phi) dp com phi=0
    integral1, _ = quad(lambda p: p * np.cos(np.radians(0)), 0, 2)
    # Integral de 2 a 0 de p*cos(phi) dp com phi=60°
    integral2, _ = quad(lambda p: p * np.cos(np.radians(60)), 2, 0)

    # Soma das integrais
    circulacao_total = integral1 + integral2
    return circulacao_total

# Calcula a circulação
circulacao_total = circulacao()
print(f"\033[1m\033[4mCirculação total\033[0m: {circulacao_total:4f} \n\n")

#-------------------------------------------------------------------------------
```

<center> <img src="https://i.imgur.com/DZkEINJ.png">

Assim como no exemplo 1.5.1 o primeiro passo é calcular nosso $\vec{F} \cdot d\vec{l}$, sabemos que agora iremos utilizar o sistema de coordenadas cartesianos, logo iremos usar o $d\vec{l}$ cartesiano, logo: <br>
$$\vec{F} \cdot d\vec{l}= (x^2 \vec{a}_x - xz\vec{a}_y - y^2\vec{a}_z) \cdot (dx \vec{a}_x + dy \vec{a}_y + dz \vec{a}_z)$$
$$\vec{F} \cdot d\vec{l}= x^2 dx - xzdy - y^2 dz$$
Pronto, agora que calculamos o produto escalar o próximo passo é entender os trajetos das integrais. Lembrando que o meu $\vec{F} \cdot \vec{dl}$ **NUNCA MUDA**. O primeiro trajeto começa no ponto $(1,0,0)$ e vai até o ponto de origem $(0,0,0)$, o segundo trajeto vai da origem $(0,0,0)$ até o $(0,1,0)$, depois vamos deste ponto $(0,1,0)$ até o $(1,1,1)$ e por fim deste ponto $(1,1,1)$ até o ponto inicial $(1,0,0)$, então teremos quatro integrais para serem resolvidas, assim como na questão anterior, se resolvermos a integral uma vez e mudarmos apenas o limite para o resto dos casos, então resolveremos a integral, sem se importar com os limites:
$$\int x^2 dx - \int xzdy - \int y^2 dz= \left[\frac{x^3}{3}\right] + [-xz(y)] + [-y^2(z)]$$ Agora basta aplicar os limites de integração, no primeiro termo varia em x, o segundo em y e o terceiro em z, então ficamos com:

*   Integral $1$
$$\int_{(1,0,0)}^{(0,0,0)} (x^2dx-xzdy-y^2dz)= \left[\frac{x^3}{3}|_{(1,0,0)}^{(0,0,0)}\right] + \left[-xz\left(y_{(1,0,0)}^{(0,0,0)}\right)\right] + \left[-y^2\left(z|_{(1,0,0)}^{(0,0,0)}\right)\right]$$
Notamos então que para o segundo e terceiro termo o y e z não irão variar, logo o resulado desses termos é 0, já para o primeiro termo ficamos com:
$$\frac{0^3}{3}-\frac{1^3}{3}=-\frac{1}{3}$$
*   Integral $2$ <Br>
Para a segunda integral ficamos com o limite de integração diferente, ficando assim: $$\int_{(0,0,0)}^{(0,1,0)} x^2dx-xzdy-y^2dz= \left[\frac{x^3}{3}|_{(0,0,0)}^{(0,1,0)}\right] + \left[-xz\left(y_{(0,0,0)}^{(0,1,0)}\right)\right] + \left[-y^2\left(z|_{(0,0,0)}^{(0,1,0)}\right)\right]$$
Para essa integral notamos que o primeiro e o terceiro termo a **variável é constante**, logo para tais termos o resultado é $0$, já para o segundo termo ficamos com:
$$\left[-xz\left(y_{(0,0,0)}^{(0,1,0)}\right)\right]= -0(1-0)=0$$
*   Integral $3$ <br>
Para a terceira integral teremos:
$$\int_{(0,1,0)}^{(1,1,1)} x^2dx-xzdy-y^2dz= \left[\frac{x^3}{3}|_{(0,1,0)}^{(1,1,1)}\right] + \left[-xz\left(y_{(0,1,0)}^{(1,1,1)}\right)\right] + \left[-y^2\left(z|_{(0,1,0)}^{(1,1,1)}\right)\right]$$
Notamos que para o termo do meio a variável $y$ não varia, logo se aplicarmos tal intervalo nela irá dar $0$, então podemos desconsiderar o termo do meio. Então temos que aplicar o intervalo no primeiro e último termo, ficando assim:
$$\left[\frac{x^3}{3}|_{(0,1,0)}^{(1,1,1)}\right] + \left[-y^2\left(z|_{(0,1,0)}^{(1,1,1)}\right)\right]=\left(\frac{1^3}{3}-\frac{0}{3}\right)+[-1^2(1-0)]$$
$$\frac{1}{3}-1=-\frac{2}{3}$$
*   Integral $4$ <br>
Para o segmento 4 temos um pequeno problema pois duas varíaveis estão variando, então temos de analisar o problema.<br>
Se olharmos bem para a figura, iremos notar que o percurso $4$ ocorre ao longo da diagonal da face mais extrema do cubo, ou seja, ele vai do ponto localizado em $(1,1,1)$ até o ponto $(1,0,0)$, então teremos o seguinte cubo: <br>
<center> <img src='https://i.imgur.com/PbDdSLC.png'> </center>
Notamos então que para qualquer ponto que olharmos ao longo da linha vermelha, que é o caminho a ser seguido, nossa variável $y$ é igual a varíavel $z$, pelo fato de ser um cubo e o trajeto ser uma diagonal, então  $y=z$ implica que $dy=dz$, então ficamos com a seguinte integral:
$$\int_{(1,1,1)}^{(1,0,0)} x^2dx-xzdy-y^2dz$$
Como nosso $x$ não varia, o primeiro a integral de $x^2dx$ é $0$, então ficamos com:
$$\int_{(1,1,1)}^{(1,0,0)}-xzdy-y^2dz$$
Como $y=z$ e $dy=dz$:
$$\int_{(1,1,1)}^{(1,0,0)}-xydy-y^2dy= x|_1\left[\int_{(1,1,1)}^{(1,0,0)}(-y-y^2)dy\right]$$
$$1\left[-\frac{y^2}{2}|_{(1,1,1)}^{(1,0,0)} - \frac{y^3}{3}|_{(1,1,1)}^{(1,0,0)}\right]=-\left(\frac{0^2}{2}-\frac{1^2}{2}\right)-\left(\frac{0^3}{3}-\frac{1^3}{3}\right)=\frac{1}{2}+\frac{1}{3}=\frac{5}{6}$$

Somando agora o valor das 4 integrais temos $-\frac{1}{3}+0-\frac{2}{3}+\frac{5}{6}=-\frac{1}{6}$.

## <big> Gradiente de um Escalar
O gradiente de um campo escalar $V$ é um vetor que representa a magnitude e a orientação da máxima taxa espacial de variação de $V$. <br>
O operador gradiente em um campo escalar $V$ é dado por $∇V$, podendo ser expresso em coordenadas cartesianas, cilíndricas e esféricas.

-   Coordenadas Cartesianas
 $$\nabla V= \frac{\partial V}{\partial x} \vec{a}_x + \frac{\partial V}{\partial y} \vec{a}_y + \frac{\partial V}{\partial z} \vec{a}_z $$

-   Coordenadas Cilíndricas
$$\nabla V= \frac{\partial V}{\partial \rho} \vec{a}_\rho + \frac{1}{\rho} \frac{\partial V}{\partial \phi} \vec{a}_\phi + \frac{\partial V}{\partial z} \vec{a}_z $$

-   Coordenadas Esféricas
 $$\nabla V= \frac{\partial V}{\partial r} \vec{a}_r + \frac{1}{r}\frac{\partial V}{\partial \theta} \vec{a}_\theta + \frac{1}{r sen\theta}\frac{\partial V}{\partial \phi} \vec{a}_\phi$$

Determine o gradiente dos seguintes campos escalares:

## <big> Divergência de um Vetor e Teorema da Divergência
Supondo um campo vetorial $\vec{A}$ temos que a divergência de $\vec{A}$ em um dado ponto *P* é o fluxo que sai, por unidade de volume, à medida que o volume se reduz à zero em torno de *P*. <br>
Assim como no gradiente, a divergência de um vetor $\vec{A}$ segue equações nas três coordenadas: <br>

-   Coordenadas Cartesianas
$$\nabla \cdot \vec{A}= \frac{\partial A_x}{\partial x}  + \frac{\partial A_y}{\partial y}  + \frac{\partial A_z}{\partial z}  $$

-   Coordenadas Cilíndricas
$$\nabla \cdot \vec{A}= \frac{1}{\rho}\frac{\partial}{\partial \rho} (\rho A_\rho) + \frac{1}{\rho} \frac{\partial A_\phi}{\partial \phi} + \frac{\partial A_z}{\partial z} $$

-   Coordenadas Esféricas
$$\nabla \cdot \vec{A}= \frac{1}{r^2}\frac{\partial}{\partial r} (r^2 A_r) + \frac{1}{rsen\theta}\frac{\partial}{\partial \theta} (A_\theta sen\theta) + \frac{1}{r sen\theta}\frac{\partial A_\phi}{\partial \phi}$$ <br>

 A partir de tais equações chegamos no **Teorema da Divergência** que estabelece que o fluxo total de um campo vetorial $\vec{A}$ que sai de uma superfície fechada *S* é igual a integral de volume da divergência de $\vec{A}$: $$\oint_S \vec{A} \cdot d\vec{S} = \int_V \nabla \cdot \vec{A} dv$$

## <big> Rotacional de um Vetor e Teorema de Stokes
O **Rotacional** de $\vec{A}$ é um vetor axial, cuja magnitude é a máxima circulação de $\vec{A}$ por unidade de área.
$$rot \vec{A} = \nabla \times \vec{A} = (\lim_{\Delta S \to 0} \frac{\oint \vec{A}.d\vec{L}}{\Delta S})_{máx} \vec{a}_n$$
<br>

Novamente a partir dessa equação podemos chegar na equação de rotacional para cada coordenada. <br>

**(Colocar também a rot de A em função do determinante para cada sistema de coordenadas)**

-   Coordenadas Cartesianas
$$ \nabla \times \vec{A}= \begin{vmatrix}
\vec{a}_x & \vec{a}_y & \vec{a}_z\\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\
A_x & A_y & A_z
\end{vmatrix} $$
<center> ou </center>
$$\nabla \times \vec{A}= [\frac{\partial A_z}{\partial y}- \frac{\partial A_y}{\partial z}] \vec{a}_x  + [\frac{\partial A_x}{\partial z}- \frac{\partial A_z}{\partial x}] \vec{a}_y  + [\frac{\partial A_y}{\partial x}- \frac{\partial A_x}{\partial y}] \vec{a}_z $$
<br>

-   Coordenadas Cilíndricas
$$ \nabla \times \vec{A}= \frac{1}{\rho}
\begin{vmatrix}
\vec{a}_\rho & \rho \vec{a}_\phi & \vec{a}_z\\
\frac{\partial}{\partial \rho} & \frac{\partial}{\partial \phi} & \frac{\partial}{\partial z} \\
A_\rho & \rho A_\phi & A_z
\end{vmatrix} $$
<center> ou </center>
$$\nabla \times \vec{A}= [\frac{1}{\rho}\frac{\partial A_z}{\partial \phi}- \frac{\partial A_\phi}{\partial z}] \vec{a}_\rho  + [\frac{\partial A_\rho}{\partial z}- \frac{\partial A_z}{\partial \rho}] \vec{a}_\phi  + \frac{1}{\rho} [\frac{\partial (\rho A_\phi)}{\partial \rho}- \frac{\partial A_\rho}{\partial \phi}] \vec{a}_z $$
<br>

-   Coordenadas Esféricas
$$ \nabla \times \vec{A}= \frac{1}{r^2 sen \theta}
\begin{vmatrix}
\vec{a}_r & r \vec{a}_\theta & r sen \theta\vec{a}_\phi\\
\frac{\partial}{\partial r} & \frac{\partial}{\partial \theta} & \frac{\partial}{\partial \phi} \\
A_r & r A_\theta & r sen\theta A_\phi
\end{vmatrix} $$
<center> ou </center>
$$\nabla \times \vec{A}= [\frac{1}{rsen\theta}\frac{\partial (A_\phi sen\theta)}{\partial \theta}- \frac{\partial A_\theta}{\partial \phi}] \vec{a}_r  + \frac{1}{r}[\frac{1}{sen\theta}\frac{\partial A_r}{\partial \phi}- \frac{\partial (r A_\phi)}{\partial r}] \vec{a}_\theta  + \frac{1}{r} [\frac{\partial (r A_\theta)}{\partial r}- \frac{\partial A_r}{\partial \theta}] \vec{a}_\phi$$
<br>

 A partir de tais equações podemos deduzir para, $$\oint_L \vec{A} \cdot d\vec{l} = \int_S (\nabla \times \vec{A}) \cdot d\vec{S}$$
 <br>

 Essa dedução é o **Teorema de Stokes** que nos diz que a circulação de um campo vetorial $\vec{A}$ em torno de um caminho fechado L é igual à integral de superfície do rotacional de $\vec{A}$ sobre a superfície aberta *S*.

### <big> Exemplo 1.6.7 <br>

Determine o rotacional de cada um dos campos vetoriais:

<big> (a) $\vec{A} = yz \vec{a}_x + 4xy\vec{a}_y + y\vec{a}_z$


```python
# @title Item (a)
from sympy import symbols, diff, sin, cos,simplify

# Definindo variáveis simbólicas
x, y, z, rho, phi, r, theta = symbols('x y z rho phi r theta')

# Definindo os campos vetoriais
Ax = y*z # Edite aqui para outra função da amplitude em x
Ay= 4*x*y # Edite aqui para outra função da amplitude em y
Az= y # Edite aqui para outra função da amplitude em z

# Calculando o rotacional de cada campo vetorial
rot_A = (
    diff(Az, y) - diff(Ay, z),
    diff(Ax, z) - diff(Az, x),
    diff(Ay, x) - diff(Ax, y)
)

# Exibindo o resultado
print(f"\033[1m\033[4mRotacional\033[0m de A: {simplify(rot_A)}")
```

## <big> Laplaciano de um Escalar
O **Laplaciano** de um campo escalar $V$, escrito como $\nabla^2V$, é o divergente do gradiente de $V$. Novamente temos nas três coordenadas suas equações. <br>


-   Coordenadas Cartesianas

$$\nabla^2 V= \frac{\partial ^2 V}{\partial x^2} + \frac{\partial ^2 V}{\partial y^2} + \frac{\partial ^2 V}{\partial z^2}$$

<br>

-   Coordenadas Cilíndricas

$$ \nabla^2 V= \frac{1}{\rho} \frac{\partial}{\partial \rho}(\rho \frac{\partial V}{\partial \rho}) + \frac{1}{\rho ^2} \frac{\partial^2 V}{\partial \phi^2} + \frac{\partial^2 V}{\partial z^2}$$
<br>

-   Coordenadas Esféricas

$$\nabla^2 V= \frac{1}{r^2} \frac{\partial}{\partial r}(r^2 \frac{\partial V}{\partial r}) + \frac{1}{r^2 sen\theta} \frac{\partial}{\partial \theta}(sen \theta \frac{\partial V}{\partial \theta}) + \frac{1}{r^2 sen^2 \theta} \frac{\partial^2 V}{\partial \phi^2}$$
<br>
